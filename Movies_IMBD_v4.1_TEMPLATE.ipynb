# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
data = pd.read_csv('movie_bd_v5.xls') 
#создаем колонку с длинами названий фильмов
data['title_len'] = data['original_title'].apply(lambda x:len(x))
data['production_companies'] = data['production_companies'].apply(lambda x:x.split('|'))
#выбираем самые длинные названия
data[data['title_len'] > data['title_len'].mean()]
production_companies_df = data.explode('production_companies')
production_companies_df = production_companies_df.groupby('production_companies').title_len.mean().sort_values(ascending=False).index[0]
print(production_companies_df)
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
data = pd.read_csv('movie_bd_v5.xls')
for item in data['overview']:
    for char in '-.,':
        item = item.replace(char, '')
        item_word_overview = item.split()
#создаем колонку с длинами описаний фильмов
data.insert(loc=0, column='len_overview', value=item_word_overview)
data['production_companies'] = data['production_companies'].apply(lambda x:x.split('|'))
#выбираем самые длинные описания
data[data['len_overview'] > data['len_overview'].mean()]
production_companies_df = data.explode('production_companies')
production_companies_df = production_companies_df.groupby('production_companies').len_overview.mean().sort_values(ascending=False).index[0]
print(production_companies_df)
```
